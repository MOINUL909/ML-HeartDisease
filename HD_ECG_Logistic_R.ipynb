Importing the Dependency

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

DataSet


```python
ecg= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ECG/heart.csv')
```

1 represents Abnormal AND 0 represent Normal

Split Features And Target

```python
x = ecg.drop(columns='target', axis=1)
y=ecg['target']
```

```python
x
```

```python
y
```

Split Train Test

```python
X_train,X_test,Y_train,Y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, X_train.shape, X_test.shape)
```

Model

Logistric Regression

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

Acurracy Score

```python
X_train_predection = model.predict(X_train)
training_Data_accuracy = accuracy_score(X_train_predection, Y_train)

```

```python
print('Accuracy Of Train data is : ', training_Data_accuracy)
```

```python
X_test_predection = model.predict(X_test)
test_Data_accuracy = accuracy_score(X_test_predection, Y_test)

```

```python
print('Accuracy Of Test data is : ', test_Data_accuracy)
```

Building a productive System

```python
input_data= (47,1,2,138,257,0,0,156,0,0,2,0,2)
input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
print(prediction)
```

```python
if(prediction[0]==0):
  print('The person is Normal. Does not Found any abnormal issues')
else:
  print("The person is not Normal. Found Heart Disease")
```
